```python
import os, numpy as np, torch, torch.nn as nn, torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from pathlib import Path
from tqdm.notebook import tqdm
import matplotlib.pyplot as plt

# ── Config ────────────────────────────────────────────────────────────────────
DATA_PATH   = '/hpcstor6/scratch01/b/brian.ngan001/spleen_synthesis/Preprocessed_Roboflow'
OUTPUT_DIR  = '/hpcstor6/scratch01/b/brian.ngan001/spleen_synthesis/ddpm_chimera_checkpoints'
RESULTS_DIR = '/hpcstor6/scratch01/b/brian.ngan001/spleen_synthesis/ddpm_chimera_results'
IMAGE_SIZE  = 256
BATCH_SIZE  = 8
LR          = 5e-5
NUM_EPOCHS  = 60
T           = 1000
SAVE_EVERY  = 5
NUM_GENERATE = 200
SEED        = 42

torch.manual_seed(SEED)
device = torch.device('cuda')
Path(OUTPUT_DIR).mkdir(parents=True, exist_ok=True)
Path(RESULTS_DIR).mkdir(parents=True, exist_ok=True)
print(f"Device: {torch.cuda.get_device_name(0)}")

# ── Dataset ───────────────────────────────────────────────────────────────────
class SpleenDataset(Dataset):
    def __init__(self, paths):
        self.paths = paths
    def __len__(self):
        return len(self.paths)
    def __getitem__(self, idx):
        d   = np.load(self.paths[idx])
        img = d['image'].astype(np.float32)
        img = img / 127.5 - 1.0
        return torch.from_numpy(img).unsqueeze(0)

all_paths = sorted([
    os.path.join(DATA_PATH, f)
    for f in os.listdir(DATA_PATH) if f.endswith('.npz')
])
np.random.shuffle(all_paths)
cut      = int(len(all_paths) * 0.9)
train_ds = SpleenDataset(all_paths[:cut])
val_ds   = SpleenDataset(all_paths[cut:])
train_dl = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True,
                      num_workers=4, pin_memory=True, drop_last=True)
val_dl   = DataLoader(val_ds, batch_size=BATCH_SIZE, shuffle=False,
                      num_workers=4, pin_memory=True)
print(f"Train: {len(train_ds)} | Val: {len(val_ds)}")

# ── U-Net (base_ch=128) ───────────────────────────────────────────────────────
class SinusoidalPE(nn.Module):
    def __init__(self, dim):
        super().__init__()
        self.dim = dim
    def forward(self, t):
        device = t.device
        half   = self.dim // 2
        emb    = torch.log(torch.tensor(10000.0)) / (half - 1)
        emb    = torch.exp(torch.arange(half, device=device) * -emb)
        emb    = t[:, None].float() * emb[None, :]
        return torch.cat([emb.sin(), emb.cos()], dim=-1)

class ResBlock(nn.Module):
    def __init__(self, in_ch, out_ch, t_dim):
        super().__init__()
        self.norm1  = nn.GroupNorm(min(8, in_ch), in_ch)
        self.conv1  = nn.Conv2d(in_ch, out_ch, 3, padding=1)
        self.norm2  = nn.GroupNorm(min(8, out_ch), out_ch)
        self.conv2  = nn.Conv2d(out_ch, out_ch, 3, padding=1)
        self.t_proj = nn.Linear(t_dim, out_ch)
        self.skip   = nn.Conv2d(in_ch, out_ch, 1) if in_ch != out_ch else nn.Identity()
        self.act    = nn.SiLU()
    def forward(self, x, t_emb):
        h = self.act(self.norm1(x))
        h = self.conv1(h)
        h = h + self.t_proj(self.act(t_emb))[:, :, None, None]
        h = self.act(self.norm2(h))
        h = self.conv2(h)
        return h + self.skip(x)

class AttentionBlock(nn.Module):
    def __init__(self, ch):
        super().__init__()
        self.norm = nn.GroupNorm(min(8, ch), ch)
        self.attn = nn.MultiheadAttention(ch, num_heads=4, batch_first=True)
    def forward(self, x):
        B, C, H, W = x.shape
        h = self.norm(x).view(B, C, H*W).transpose(1, 2)
        h, _ = self.attn(h, h, h)
        return x + h.transpose(1, 2).view(B, C, H, W)

class UNet(nn.Module):
    def __init__(self, in_ch=1, base_ch=128, t_dim=256):
        super().__init__()
        self.t_emb = nn.Sequential(
            SinusoidalPE(t_dim),
            nn.Linear(t_dim, t_dim * 4),
            nn.SiLU(),
            nn.Linear(t_dim * 4, t_dim),
        )
        self.enc1  = ResBlock(in_ch,      base_ch,    t_dim)
        self.enc2  = ResBlock(base_ch,    base_ch*2,  t_dim)
        self.enc3  = ResBlock(base_ch*2,  base_ch*4,  t_dim)
        self.enc4  = ResBlock(base_ch*4,  base_ch*8,  t_dim)
        self.down  = nn.MaxPool2d(2)
        self.mid1  = ResBlock(base_ch*8,  base_ch*8,  t_dim)
        self.attn  = AttentionBlock(base_ch*8)
        self.mid2  = ResBlock(base_ch*8,  base_ch*8,  t_dim)
        self.up4   = nn.ConvTranspose2d(base_ch*8, base_ch*8, 2, stride=2)
        self.dec4  = ResBlock(base_ch*16, base_ch*4,  t_dim)
        self.up3   = nn.ConvTranspose2d(base_ch*4, base_ch*4, 2, stride=2)
        self.dec3  = ResBlock(base_ch*8,  base_ch*2,  t_dim)
        self.up2   = nn.ConvTranspose2d(base_ch*2, base_ch*2, 2, stride=2)
        self.dec2  = ResBlock(base_ch*4,  base_ch,    t_dim)
        self.up1   = nn.ConvTranspose2d(base_ch,   base_ch,   2, stride=2)
        self.dec1  = ResBlock(base_ch*2,  base_ch,    t_dim)
        self.out   = nn.Sequential(
            nn.GroupNorm(min(8, base_ch), base_ch),
            nn.SiLU(),
            nn.Conv2d(base_ch, in_ch, 1),
        )
    def forward(self, x, t):
        t_emb = self.t_emb(t)
        e1 = self.enc1(x,             t_emb)
        e2 = self.enc2(self.down(e1), t_emb)
        e3 = self.enc3(self.down(e2), t_emb)
        e4 = self.enc4(self.down(e3), t_emb)
        m  = self.mid1(self.down(e4), t_emb)
        m  = self.attn(m)
        m  = self.mid2(m,             t_emb)
        d4 = self.dec4(torch.cat([self.up4(m),  e4], dim=1), t_emb)
        d3 = self.dec3(torch.cat([self.up3(d4), e3], dim=1), t_emb)
        d2 = self.dec2(torch.cat([self.up2(d3), e2], dim=1), t_emb)
        d1 = self.dec1(torch.cat([self.up1(d2), e1], dim=1), t_emb)
        return self.out(d1)

# ── DDPM Scheduler ────────────────────────────────────────────────────────────
class DDPMScheduler:
    def __init__(self, T=1000, beta_start=1e-4, beta_end=0.02):
        self.T        = T
        betas         = torch.linspace(beta_start, beta_end, T)
        alphas        = 1.0 - betas
        alpha_cumprod = torch.cumprod(alphas, dim=0)
        self.betas         = betas.to(device)
        self.alphas        = alphas.to(device)
        self.alpha_cumprod = alpha_cumprod.to(device)
        self.sqrt_acp      = torch.sqrt(alpha_cumprod).to(device)
        self.sqrt_1m_acp   = torch.sqrt(1 - alpha_cumprod).to(device)

    def add_noise(self, x0, t, noise=None):
        if noise is None:
            noise = torch.randn_like(x0)
        sqrt_acp   = self.sqrt_acp[t][:, None, None, None]
        sqrt_1macp = self.sqrt_1m_acp[t][:, None, None, None]
        return sqrt_acp * x0 + sqrt_1macp * noise, noise

    @torch.no_grad()
    def sample(self, model, n=1, img_size=256):
        model.eval()
        x = torch.randn(n, 1, img_size, img_size, device=device)
        for t in tqdm(reversed(range(self.T)), desc='Sampling', total=self.T, leave=False):
            t_batch    = torch.full((n,), t, device=device, dtype=torch.long)
            with torch.amp.autocast('cuda'):
                noise_pred = model(x, t_batch)
            beta_t  = self.betas[t]
            alpha_t = self.alphas[t]
            acp_t   = self.alpha_cumprod[t]
            x = (1 / torch.sqrt(alpha_t)) * (
                x - (beta_t / torch.sqrt(1 - acp_t)) * noise_pred
            )
            if t > 0:
                x += torch.sqrt(beta_t) * torch.randn_like(x)
        return x.clamp(-1, 1)

# ── Build model ───────────────────────────────────────────────────────────────
model     = UNet(in_ch=1, base_ch=128, t_dim=256).to(device)
scheduler = DDPMScheduler(T=T)
n_params  = sum(p.numel() for p in model.parameters())
print(f"DDPM U-Net parameters: {n_params/1e6:.1f}M")

optimizer = torch.optim.AdamW(model.parameters(), lr=LR, weight_decay=1e-4)
lr_sched  = torch.optim.lr_scheduler.CosineAnnealingLR(
    optimizer, T_max=len(train_dl) * NUM_EPOCHS)
scaler    = torch.amp.GradScaler('cuda')

# ── Training ──────────────────────────────────────────────────────────────────
print(f"\nStarting DDPM training for {NUM_EPOCHS} epochs...")
best_val     = float('inf')
train_losses = []
val_losses   = []

for epoch in range(NUM_EPOCHS):
    model.train()
    ep_loss = 0.0

    for imgs in tqdm(train_dl, desc=f'Epoch {epoch+1}/{NUM_EPOCHS}'):
        imgs = imgs.to(device)
        B    = imgs.size(0)
        t    = torch.randint(0, T, (B,), device=device).long()

        with torch.amp.autocast('cuda'):
            noisy, noise = scheduler.add_noise(imgs, t)
            pred         = model(noisy, t)
            loss         = F.mse_loss(pred, noise)

        scaler.scale(loss).backward()
        scaler.unscale_(optimizer)
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        scaler.step(optimizer)
        scaler.update()
        optimizer.zero_grad()
        lr_sched.step()
        ep_loss += loss.item()

    avg_train = ep_loss / len(train_dl)
    train_losses.append(avg_train)

    # Validation
    model.eval()
    v_loss = 0.0
    with torch.no_grad():
        for imgs in val_dl:
            imgs = imgs.to(device)
            B    = imgs.size(0)
            t    = torch.randint(0, T, (B,), device=device).long()
            with torch.amp.autocast('cuda'):
                noisy, noise = scheduler.add_noise(imgs, t)
                pred         = model(noisy, t)
                v_loss      += F.mse_loss(pred, noise).item()

    avg_val = v_loss / len(val_dl)
    val_losses.append(avg_val)
    print(f'Epoch {epoch+1:3d} | train={avg_train:.4f} | val={avg_val:.4f}')

    if (epoch + 1) % SAVE_EVERY == 0:
        torch.save({
            'epoch'     : epoch,
            'model'     : model.state_dict(),
            'optimizer' : optimizer.state_dict(),
            'val_loss'  : avg_val,
        }, f'{OUTPUT_DIR}/epoch-{epoch+1:03d}.pt')
        print(f'Checkpoint saved → epoch-{epoch+1:03d}.pt')

    if avg_val < best_val:
        best_val = avg_val
        torch.save(model.state_dict(), f'{OUTPUT_DIR}/best.pt')
        print(f'Best model updated (val={best_val:.4f})')

    # Progress samples every 10 epochs
    if (epoch + 1) % 10 == 0:
        samples = scheduler.sample(model, n=4, img_size=IMAGE_SIZE)
        samples = (samples + 1) / 2
        fig, axes = plt.subplots(1, 4, figsize=(16, 4))
        fig.suptitle(f'Generated samples — epoch {epoch+1}')
        for i, ax in enumerate(axes):
            ax.imshow(samples[i, 0].cpu().numpy(), cmap='gray')
            ax.axis('off')
        plt.tight_layout()
        plt.savefig(f'{RESULTS_DIR}/samples_epoch{epoch+1:03d}.png', dpi=120, bbox_inches='tight')
        plt.show()
        model.train()

print('\nTraining complete!')

# ── Loss curve ────────────────────────────────────────────────────────────────
plt.figure(figsize=(10, 4))
plt.plot(train_losses, label='Train')
plt.plot(val_losses,   label='Val')
plt.xlabel('Epoch'); plt.ylabel('MSE Loss')
plt.title('DDPM Training Loss')
plt.legend(); plt.grid(True)
plt.savefig(f'{RESULTS_DIR}/loss_curve.png', dpi=150, bbox_inches='tight')
plt.show()

# ── Final generation ──────────────────────────────────────────────────────────
print(f'Generating {NUM_GENERATE} synthetic images...')
all_synth = []
for i in tqdm(range(0, NUM_GENERATE, 4), desc='Generating'):
    batch = min(4, NUM_GENERATE - i)
    imgs  = scheduler.sample(model, n=batch, img_size=IMAGE_SIZE)
    imgs  = (imgs + 1) / 2
    for img in imgs:
        all_synth.append((img[0].cpu().numpy() * 255).astype(np.uint8))

np.savez_compressed(f'{RESULTS_DIR}/ddpm_synthetic.npz', images=np.stack(all_synth))
print(f'Saved {len(all_synth)} synthetic images.')

# ── Final comparison grid ─────────────────────────────────────────────────────
real_samples = []
for p in all_paths[:8]:
    d = np.load(p)
    real_samples.append(d['image'])

fig, axes = plt.subplots(2, 8, figsize=(20, 6))
fig.suptitle('Top: Real spleen US    |    Bottom: DDPM synthetic',
             fontsize=14, fontweight='bold')
for i in range(8):
    axes[0, i].imshow(real_samples[i], cmap='gray')
    axes[0, i].axis('off')
    axes[0, i].set_title('Real', fontsize=9)
    axes[1, i].imshow(all_synth[i], cmap='gray')
    axes[1, i].axis('off')
    axes[1, i].set_title('Synthetic', fontsize=9)
plt.tight_layout()
plt.savefig(f'{RESULTS_DIR}/real_vs_synthetic.png', dpi=150, bbox_inches='tight')
plt.show()
print('All done!')
```
